```python
import pandas as pd
```

```python
filename = '/content/mortgages.csv'
df = pd.read_csv(filename)
```

Rename columns

```python
df.head()
```

```python
df.Principal Paid
```

```python
df = df.rename(columns={'Starting Balance': 'starting_balance',
                        'Interest Paid': 'interest_paid', 
                        'Principal Paid': 'principal_paid'})
df.head()
```

```python
df.columns = ['month',
              'starting_balance',
              'repayment',
              'interest_paid',
              'principal_paid',
              'new_balance',
              'mortgage_name',
              'interest_rate']
df.head()
```

Delete Columns

```python
df = df.drop(columns=['new_balance'])
df.head()
```

```python
del df['starting_balance']
df.head()
```
